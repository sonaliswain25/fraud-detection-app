```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
df.head()
```

```python
df_model = df.drop (["nameOrig" , "nameDest" , "isFlaggedFraud"], axis = 1)
```

```python
df_model.head()
```

```python
categorical = ["type"]
numeric = ["amount","oldbalanceOrg","newbalanceOrig","oldbalanceDest","newbalanceDest"]
```

```python
y = df_model["isFraud"]
x = df_model.drop("isFraud",axis =1)
```

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, stratify = y)
```

```python
preprocessor = ColumnTransformer(
      transformers= [
 		("num",  StandardScaler(), numeric),
        ("cat", OneHotEncoder(drop= "first"), categorical)
    ],
      remainder = "drop"
)
```

```python
pipeline = Pipeline([
("prep", preprocessor),
("clf", LogisticRegression(class_weight= "balanced" , max_iter=1000))
])
```

```python
pipeline.fit(x_train, y_train)
```

```python
y_pred_lr = pipeline.predict(x_test)

lr_accuracy = accuracy_score(y_test, y_pred_lr)

print("Logistic Regression Accuracy:", lr_accuracy)
print("\nClassification Report:\n", classification_report(y_test, y_pred_lr))
```

```python
confusion_matrix(y_test, y_pred_lr)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve,auc
import matplotlib.pyplot as pltc
```

```python
# Logistic Regression
y_proba_lr = pipeline.predict_proba(x_test)[:, 1]
roc_lr = roc_auc_score(y_test, y_proba_lr)
print("Logistic ROC-AUC:", round(roc_lr, 4))
```

```python
#ROC Curve for Logistic Regression
fpr, tpr, thresholds = roc_curve(y_test, y_proba_lr)
```

```python
plt.figure()
plt.plot(fpr, tpr, label=f"AUC = {roc_lr:.4f}")
plt.plot([0, 1], [0, 1])  # Random classifier line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Fraud Detection Model")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
pipeline.score(x_test, y_test) * 100
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
categorical_cols = ["type"]
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols)
    ],
    remainder="passthrough"
)
```

```python
rf_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(n_estimators=50, random_state=42))
])
```

```python
rf_pipeline.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred_rf = rf_pipeline.predict(x_test)

rf_accuracy = accuracy_score(y_test, y_pred_rf)

print("Random Forest Accuracy:", rf_accuracy)
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))
```

```python
confusion_matrix(y_test, y_pred_rf)

```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
```

```python
#ROC Curve for Random Forest
y_proba_rf = rf_pipeline.predict_proba(x_test)[:, 1]
roc_rf = roc_auc_score(y_test, y_proba_rf)

print("Random Forest ROC-AUC:", round(roc_rf, 4))
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_proba_rf)
```

```python

plt.figure()
plt.plot(fpr, tpr, label=f"Random Forest AUC = {roc_rf:.4f}")
plt.plot([0, 1], [0, 1])  # Diagonal line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Random Forest")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
rf_pipeline.score(x_test, y_test) * 100
```

```python
results = pd.DataFrame({
    "Model": ["Logistic Regression", "Random Forest"],
    "Accuracy": [accuracy_score(y_test, y_pred_rf), rf_accuracy],
    "ROC-AUC": [roc_lr, roc_rf]
})

print(results)
```

```python
# Get feature names after encoding
feature_names = rf_pipeline.named_steps["preprocessor"].get_feature_names_out()

# Get importance
importances = rf_pipeline.named_steps["model"].feature_importances_

# Create DataFrame
feat_df = pd.DataFrame({
    "Feature": feature_names,
    "Importance": importances
}).sort_values(by="Importance", ascending=False)

# Plot
plt.figure(figsize=(10,5))
plt.bar(feat_df["Feature"], feat_df["Importance"])
plt.xticks(rotation=90)
plt.title("Feature Importance - Random Forest")
plt.xlabel("Features")
plt.ylabel("Importance")
plt.show()
```
